```python
import pandas as pd
from sklearn.model_selection import train_test_split
import os

# --- 1. Define Correct Column Names ---
LABEL_COLUMN = 'image_type'
FILENAME_COLUMN = 'fname'

# --- 2. Load the CSV ---
DATA_DIR = 'tbx11k-simplified'
csv_path = f'{DATA_DIR}/data.csv'
df = pd.read_csv(csv_path)

# --- 3. Create Full File Paths ---
def get_full_path(filename):
    path1 = os.path.join(DATA_DIR, 'images', filename)
    path2 = os.path.join(DATA_DIR, 'test', filename)

    if os.path.exists(path1):
        return path1
    elif os.path.exists(path2):
        return path2
    else:
        return None

df['full_path'] = df[FILENAME_COLUMN].apply(get_full_path)
df = df.dropna(subset=['full_path'])
print(f"Found {len(df)} total image files.")

# --- 4. NEW: Balance Dataset to match Paper's 3600 Images ---
# We will sample 1200 images from each class, as shown in Table 1
TARGET_CLASS_SIZE = 1200
print(f"\nBalancing dataset: Sampling {TARGET_CLASS_SIZE} images from each class.")

# Group by label, sample 1200 from each
balanced_df = df.groupby(LABEL_COLUMN, group_keys=False).apply(
    lambda x: x.sample(TARGET_CLASS_SIZE, random_state=42)
)

# Shuffle the new balanced DataFrame
balanced_df = balanced_df.sample(frac=1, random_state=42).reset_index(drop=True)

print(f"\n--- New Balanced Class Distribution ---")
print(balanced_df[LABEL_COLUMN].value_counts())

# --- 5. Split to match Paper's 2880 / 360 / 360 split ---
# We use test_size=0.2 to get 3600 * 0.2 = 720 (for val+test)
# and 3600 - 720 = 2880 (for train)
train_df, temp_test_df = train_test_split(
    balanced_df,
    test_size=0.20, # 20% of 3600 is 720
    random_state=42,
    stratify=balanced_df[LABEL_COLUMN]
)

# Now split the 720 images 50/50 to get 360 for val and 360 for test
val_df, test_df = train_test_split(
    temp_test_df,
    test_size=0.50, # 50% of 720 is 360
    random_state=42,
    stratify=temp_test_df[LABEL_COLUMN]
)

print(f"\n--- Data Split (to match Paper) ---")
print(f"Total balanced images: {len(balanced_df)}") # Should be 3600
print(f"Training images:   {len(train_df)}")   # Should be 2880
print(f"Validation images: {len(val_df)}") # Should be 360
print(f"Test images:       {len(test_df)}")   # Should be 360
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# --- 1. Define Constants ---
IMAGE_SIZE = 224
BATCH_SIZE = 32
# This must match the column name from the previous step
LABEL_COLUMN = 'image_type'

# --- 2. Augmentation for the Training Set ---
# These are the settings from the paper [cite: 240]
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=30,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.25,
    zoom_range=0.2,
    horizontal_flip=True
)

# --- 3. Only Rescaling for Validation and Testing ---
val_test_datagen = ImageDataGenerator(rescale=1./255)

# --- 4. Create the Generators from DataFrame ---
print("\nCreating generators...")
train_generator = train_datagen.flow_from_dataframe(
    dataframe=train_df,
    x_col='full_path',           # Column with full file paths
    y_col=LABEL_COLUMN,          # Column with labels
    target_size=(IMAGE_SIZE, IMAGE_SIZE),
    batch_size=BATCH_SIZE,
    class_mode='categorical'     # 3 classes: healthy, sick_but_no_tb, tb
)

validation_generator = val_test_datagen.flow_from_dataframe(
    dataframe=val_df,
    x_col='full_path',
    y_col=LABEL_COLUMN,
    target_size=(IMAGE_SIZE, IMAGE_SIZE),
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    shuffle=False                # No need to shuffle val/test
)

test_generator = val_test_datagen.flow_from_dataframe(
    dataframe=test_df,
    x_col='full_path',
    y_col=LABEL_COLUMN,
    target_size=(IMAGE_SIZE, IMAGE_SIZE),
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    shuffle=False
)

print(f"Classes found and mapped: {train_generator.class_indices}")
```

```python
import tensorflow as tf
from tensorflow.keras.layers import (
    Input, Conv2D, Dense, GlobalAveragePooling2D, Reshape, Add, LayerNormalization,
    MultiHeadAttention, Dropout, Embedding, BatchNormalization, Activation
)
from tensorflow.keras.models import Model

# --- 1. Define Hyperparameters for DeiT-T (from Table 2) ---
PATCH_SIZE = 16
NUM_PATCHES = (IMAGE_SIZE // PATCH_SIZE) ** 2
EMBED_DIM = 192    # DeiT-T Embedding size [cite: 262]
NUM_HEADS = 3      # DeiT-T Heads [cite: 262]
NUM_TRANSFORMER_LAYERS = 12 # DeiT-T Layers [cite: 262]
MLP_DIM = EMBED_DIM * 4
DROPOUT_RATE = 0.2    # From paper's settings [cite: 244]

# --- 2. Helper: Transformer Encoder Block ---
def transformer_encoder(x, embed_dim, num_heads, mlp_dim, dropout_rate):
    """Implements one Transformer Encoder block."""
    # Multi-Head Attention
    attn_input = x
    x = LayerNormalization(epsilon=1e-6)(x)
    x = MultiHeadAttention(
        num_heads=num_heads, key_dim=embed_dim // num_heads, dropout=dropout_rate
    )(x, x)
    x = Add()([x, attn_input]) # Residual connection

    # Feed-Forward Network (MLP)
    ffn_input = x
    x = LayerNormalization(epsilon=1e-6)(x)
    x = Dense(mlp_dim, activation=tf.nn.gelu)(x) # Paper confirms GELU [cite: 576]
    x = Dense(embed_dim)(x)
    x = Dropout(dropout_rate)(x)
    x = Add()([x, ffn_input]) # Residual connection
    return x

# --- 3. Helper: ResNet-16 Residual Block (from Figure 1) ---
def residual_block(x, filters):
    """Implements the residual block from Figure 1."""
    shortcut = x
    # The diagram shows 1x1 -> 3x3 -> 1x1, but the "Details" box
    # only shows 1x1 -> 3x3. We'll follow the detailed diagram.
    f1, f2 = filters

    x = Conv2D(f1, (1, 1), padding='same')(x)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)

    x = Conv2D(f2, (3, 3), padding='same')(x)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)

    # Project shortcut to match filter dimensions if necessary
    if shortcut.shape[-1] != f2:
        shortcut = Conv2D(f2, (1, 1), padding='same')(shortcut)
        shortcut = BatchNormalization()(shortcut)

    x = Add()([x, shortcut])
    x = Activation('relu')(x)
    return x

# --- 4. Build the Full Model (Based on Figure 1) ---
def build_model():
    """Builds the complete DeiT-T + ResNet-16 model."""
    # === Input ===
    inputs = Input(shape=(IMAGE_SIZE, IMAGE_SIZE, 3))

    # === Part 1: Transformer (DeiT-T) ===
    # 1. Create Patches (Conv2D is an efficient patch extractor)
    x = Conv2D(
        filters=EMBED_DIM,
        kernel_size=PATCH_SIZE,
        strides=PATCH_SIZE,
        padding='valid',
        name='patch_extractor'
    )(inputs)

    # 2. Reshape to sequence of tokens
    x = Reshape((NUM_PATCHES, EMBED_DIM))(x)

    # 3. Add Positional Embeddings
    pos_embedding = Embedding(input_dim=NUM_PATCHES, output_dim=EMBED_DIM)
    positions = tf.range(start=0, limit=NUM_PATCHES, delta=1)
    x = x + pos_embedding(positions)

    # 4. Transformer Encoder Layers (DeiT-T) [cite: 140, 262]
    for _ in range(NUM_TRANSFORMER_LAYERS):
        x = transformer_encoder(x, EMBED_DIM, NUM_HEADS, MLP_DIM, DROPOUT_RATE)

    # === Part 2: CNN (ResNet-16) ===
    # Reshape tokens back into a 2D feature map
    feature_map_dim = IMAGE_SIZE // PATCH_SIZE # 14
    x_cnn = Reshape((feature_map_dim, feature_map_dim, EMBED_DIM))(x)

    # Conv block from Figure 1 ("2X 2 Conv block 16x16x512")
    # We'll use a 3x3 Conv to get to 512 filters
    x_cnn = Conv2D(512, (3, 3), padding='same', activation='relu', name='cnn_entry_conv')(x_cnn)

    # Residual Blocks from Figure 1 [cite: 183, 185, 188]
    # Filter numbers are not specified, so we'll use reasonable values
    x_cnn = residual_block(x_cnn, filters=(256, 512)) # Residual Block 1
    x_cnn = residual_block(x_cnn, filters=(256, 512)) # Residual Block 2
    x_cnn = residual_block(x_cnn, filters=(256, 512)) # Residual Block 3

    # === Part 3: Classifier Head ===
    # Global Average Pooling [cite: 191]
    x_gap = GlobalAveragePooling2D()(x_cnn)

    # Classifier Head (from Ablation Study [cite: 581-582])
    # "two dense layers...Gelu activation function and batch normalization" [cite: 581]
    x = Dense(256)(x_gap)
    x = BatchNormalization()(x)
    x = Activation('gelu')(x) # Use GELU

    x = Dense(128)(x)
    x = BatchNormalization()(x)
    x = Activation('gelu')(x)

    # Final Output Layer (3 classes) [cite: 193]
    outputs = Dense(3, activation='softmax')(x)

    model = Model(inputs=inputs, outputs=outputs)
    return model

# --- 5. Instantiate and Summarize ---
model = build_model()
model.summary()

print("\nModel built successfully.")
print(f"The paper's DeiT-T+ResNet-16 model has 6.9M parameters[cite: 272].")
print("Your model's parameter count should be in a similar range.")
```

```python
# Compile the model
# Using Adam optimizer and 0.001 learning rate, as per the paper [cite: 244]
optimizer = tf.keras.optimizers.Adam(learning_rate=0.001)

model.compile(
    optimizer=optimizer,
    loss='categorical_crossentropy',
    metrics=[
        'accuracy',
        tf.keras.metrics.Precision(name='precision'),
        # The paper calls 'Recall' 'Sensitivity' [cite: 251]
        tf.keras.metrics.Recall(name='sensitivity')
    ]
)

# Train the model
# The paper trained for 15 epochs [cite: 244]
EPOCHS = 15

print("\n--- Starting Model Training ---")
history = model.fit(
    train_generator,
    epochs=EPOCHS,
    validation_data=validation_generator,
    # Callbacks to save the best model and stop early if it stops improving
    callbacks=[
        tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True),
        tf.keras.callbacks.ModelCheckpoint('best_model.h5', monitor='val_accuracy', save_best_only=True)
    ]
)
print("--- Model Training Complete ---")
```
